```python
import matplotlib.pyplot as plt
import timeit
import decimal
import numpy as np
decimal.getcontext().prec = 100

aMin, bMin = 0, 0
aMax = 15
bMax = 200
Proba = 0.99
Probb = 0.3
Probd = 0.5
D = [22, 27, 26, 32, 31, 25, 35, 26, 28, 23]
```

```python
# P(d1, d2,..., dn)
def Pdeno(n):
    result = 0
    for a in range(aMax+1):
        for b in range(bMax+1):
            result += Pnume(n, a, b)
    return result


# P(d1|a,b)*P(d2|a,b)*...*P(dn|a,b)P(a)P(b)
def Pnume(n, a, b):
    element1 = 1/(aMax - aMin + 1)
    element2 = 1/(bMax - bMin + 1)
    element3 = 1
    for i in range(n):
        element3 *= Pdna(D[i], a, b)
    return element1 * element2 * element3


# P(dn | a, b)
def Pdna(dn, a, b):
    element = 0
    for cn in range(a+b+1):
        element += Pcn(cn, a, b) * Pdn(dn, cn)
    return element


# P(dn | cn)
def Pdn(dn, cn):
    if dn > 2*cn or dn < cn:
        return 0
    p = Probd
    element1 = factorial(cn)/ (factorial(dn-cn)*factorial(cn-(dn-cn)))
    element2 = p**(dn-cn)
    element3 = (1-p)**(cn-(dn-cn))
    return element1 * element2 * element3


# P(cn | a, b)
def Pcn(cn, a, b):
    pcn = 0
    for xa in range(a+1):
        if xa > cn:
            break
        if cn-xa > b:
            continue
        pcn += Pxaxb(xa, cn-xa, a, b)
    return pcn


# P(xa, xb | a, b)
def Pxaxb(xa, xb, a, b):
    return P("a", xa, a) * P("b", xb, b)


# P(xa | a)
def P(name, xa, a):
    p = Probb
    if name =="a": p = Proba
    element1 = factorial(a)/ (factorial(xa)*factorial(a-xa))
    element1 = decimal.Decimal(element1)
    #element2 = p**(xa)
    element2 = decimal.Decimal(p)**decimal.Decimal(xa)
    #element3 = (1-p)**(a-xa)
    element3 = (decimal.Decimal(1)-decimal.Decimal(p))**(decimal.Decimal(a)-decimal.Decimal(xa))
    return float(element1 * element2 * element3)


facTable = [1, 1] # Store computed factorial for reusing
def factorial(n):
    if n < len(facTable):
        return facTable[n]

    num = facTable[-1]
    for i in range(len(facTable), n + 1):
        num *= i
        facTable.append(num)
    return num
```

```python
postsA = []
postsB = []
time0 = timeit.default_timer()
start = timeit.default_timer()

for n in range(1, 11):
    print("n=: ", n)
    pdeno_val = Pdeno(n)

    postsA.append([])
    for a_val in range(aMax+1):
        posterior_val = 0
        for b_val in range(bMax+1):
            posterior_val += Pnume(n, a_val, b_val) / pdeno_val
        print("posterior for a=%f, %f"%(a_val, posterior_val))
        postsA[n-1].append(posterior_val)
    print("Sum of all the posterior for n=%i: %f"%(n,sum(postsA[n-1])))
    stop = timeit.default_timer()
    print('Time for this round: ', stop - start)
    start = stop
    fig, ax = plt.subplots()
    ax.plot(postsA[n-1])
    plt.show()

    postsB.append([])
    for b_val in range(bMax+1):
        posterior_val = 0
        for a_val in range(aMax+1):
            posterior_val += Pnume(n, a_val, b_val) / pdeno_val
        print("posterior for b=%f, %f"%(b_val, posterior_val))
        postsB[n-1].append(posterior_val)
    print("Sum of all the posterior for n=%i: %f"%(n,sum(postsB[n-1])))
    stop = timeit.default_timer()
    print('Time for this n round: ', stop - start)
    start = stop
    fig, ax = plt.subplots()
    ax.plot(postsB[n-1])
    plt.show()


fig, ax = plt.subplots()
for n in range(1, 11):
    ax.plot(postsA[n-1], label="n=%i"%(n))
ax.legend()
plt.ylabel('Probabilities')
plt.xlabel('a') 
plt.show()

fig, ax = plt.subplots()
for n in range(1, 11):
    ax.plot(postsB[n-1], label="n=%i"%(n))
ax.legend()
plt.ylabel('Probabilities')
plt.xlabel('b')
plt.show()
```
