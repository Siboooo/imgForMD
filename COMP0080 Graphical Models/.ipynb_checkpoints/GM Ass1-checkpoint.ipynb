```python
import matplotlib.pyplot as plt
import decimal
decimal.getcontext().prec = 100

aMin, bMin = 0, 0
aMax = 15
bMax = 200
Proba = 0.99
Probb = 0.3
Probd = 0.5
D = [22, 27, 26, 32, 31, 25, 35, 26, 28, 23]
```

```python
def posterior(n, a, b, Pd1n_temp, name):
    Pd1n_val = Pd1n_temp
    if Pd1n_val == -1:
        Pd1n_val = Pd1n(n, a, b, name)
    return Pd1na(n, a, b, name) / Pd1n_val, Pd1n_val

# P(d1..n) = sum( P(d1..n, a) ) = sum( P(d1..n | a) * P(a) )
def Pd1n(n, a, b, name):
    Max = aMax
    if name == "b": Max = bMax
    element = 0
    for i in range(Max+1):
        #print(i)
        element += Pd1na(n, i, b, name)
    return element

# P(d1..n | a) * P(a)
def Pd1na(n, a, b, name):
    element1 = 1/(aMax - aMin + 1)
    if name == "b": element1 = 1/(bMax - bMin + 1)
    element2 = 1
    for i in range(n):
        element2 = element2 * Pdna(D[i], a, b, name)
    return element1 * element2

# P(dn | a)
def Pdna(dn, a, b, name):
    element1 = aMax - aMin + 1
    if name == "b": element1 = bMax - bMin + 1
    element2 = 0
    for b_val in range(b+1):
        for cn in range(b+a+1):
            element2 += (1/(aMax - aMin + 1)) * (1/(bMax - bMin + 1)) * (Pcn(cn, a, b_val, name)) * (Pdn(dn, cn, name))
    return element1 * element2

# P(dn | cn)
def Pdn(dn, cn, name):
    if dn > 2*cn or dn < cn:
        return 0
    p = Probd
    element1 = factorial(cn)/ (factorial(dn-cn)*factorial(cn-(dn-cn)))
    element2 = p**(dn-cn)
    element3 = (1-p)**(cn-(dn-cn))
    return element1 * element2 * element3

# P(cn | a, b)
def Pcn(cn, a, b, name):
    pcn = 0
    for xa in range(a+1):
        if xa > cn:
            break
        pcn += Pxaxb(xa, cn-xa, a, b, name)
    return pcn

# P(xa, xb | a, b)
def Pxaxb(xa, xb, a, b, name):
    if name == "b":
        return P("b", xa, a) * P("a", xb, b)
    else:
        return P("a", xa, a) * P("b", xb, b)

# P(xa | a)
def P(name, xa, a):
    p = Probb
    if name =="a": p = Proba
    element1 = factorial(a)/ (factorial(xa)*factorial(a-xa))
    element1 = decimal.Decimal(element1)
    #element2 = p**(xa)
    element2 = decimal.Decimal(p)**decimal.Decimal(xa)
    #element3 = (1-p)**(a-xa)
    element3 = (decimal.Decimal(1)-decimal.Decimal(p))**(decimal.Decimal(a)-decimal.Decimal(xa))
    return float(element1 * element2 * element3)

def factorial(n):
    x = 1
    for i in range(1, n + 1):
        x = x * i
    return x
```

```python
postsA = []

for n in range(1, 11):
    postsA.append([])
    print("n=: ", n)
    pd1n_val = -1
    for a_val in range(aMax+1):
        posterior_val, pd1n_val = posterior(n, a_val, bMax, pd1n_val, "a")
        print("posterior for a=%f, %f"%(a_val, posterior_val))
        postsA[n-1].append(posterior_val)
        
    print("Sum of all the posterior for n=%i: %f"%(n,sum(postsA[n-1])))
    fig, ax = plt.subplots()
    ax.plot(postsA[n-1])
    plt.show()

fig, ax = plt.subplots()
for n in range(1, 11):
    ax.plot(postsA[n-1], label="n=%i"%(n-1))
ax.legend()
plt.ylabel('Probabilities')
plt.xlabel('a')
plt.show()
```

```python
postsB = []

for n in range(1, 11):
    postsB.append([])
    print("n=: ", n)
    pd1n_val = -1
    for b_val in range(bMax+1):
        posterior_val, pd1n_val = posterior(n, b_val, aMax, pd1n_val, "b")
        print("posterior for b=%f, %f"%(b_val, posterior_val))
        postsB[n-1].append(posterior_val)
        
    print("Sum of all the posterior for n=%i: %f"%(n,sum(postsB[n-1])))
    fig, ax = plt.subplots()
    ax.plot(postsB[n-1])
    plt.show()

fig, ax = plt.subplots()
for n in range(1, 11):
    ax.plot(postsB[n-1], label="n=%i"%(n-1))
ax.legend()
plt.ylabel('Probabilities')
plt.xlabel('b')
plt.show()
```
